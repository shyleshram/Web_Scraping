```python
from bs4 import BeautifulSoup as soup
import requests
```

```python
my_url = 'https://www.newegg.com/global/in-en/p/pl?d=graphics+carrd'
```

```python
source = requests.get(my_url).text
```

```python
page_soup = soup(source, 'lxml')
```

```python
print(page_soup.prettify())
```

```python
print(page_soup.h1)
print(page_soup.p)
```

```python
containers = page_soup.find_all('div', class_='item-container')
len(containers)
```

```python
containers[0]
```

```python
filename = "graphics_card_product.csv"
f = open(filename, "w")

headers = "brand, product_name, price\n"

f.write(headers)
```

```python
for container in page_soup.find_all('div', class_='item-container'):
    brand_name = container.find('a', class_='item-brand')
    brand_name = brand_name.img['title']
    print(brand_name)
    
    product_name = container.find('a', class_='item-title').text
    print(product_name)
    
    price = container.find('li', class_='price-current').text.strip().split(" ")[1][:6]
    print(price)
    
    f.write(brand_name + "," + product_name.replace(",","|") + "," + price.replace(",","") + "\n")
    
f.close()
```
